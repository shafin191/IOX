```python
import os
import glob
import pandas as pd

root_directory = '/disk/mnemo/users/randomuser/SBERT All Embedding/Fake_Tweet_Still_2024/'

df_list = []

csv_files = glob.glob(os.path.join(root_directory, '**', '*.csv'), recursive=True)

for file in csv_files:
    try:
        if os.path.getsize(file) > 0:
            df = pd.read_csv(file)
            
            df['file_name'] = os.path.basename(file)
            
            df_list.append(df)
        else:
            print(f"Skipped empty file: {file}")
    except pd.errors.EmptyDataError:
        print(f"Error reading file: {file}. It might be empty or malformed.")
    
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    
    print(combined_df)
else:
    print("No valid CSV files found.")
```
